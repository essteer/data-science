# K-Means Clustering Demo

A simple example of plotting pre-clustered random data using scikit-learn.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
# Set plot scheme
palette = {"dark": "#252b33", "grey": "#45464d", "light": "#fefeff", "stone": "#8f8f94",
           "blue": "#336681", "green": "#089389", "red": "#d34748", "pink": "#cf82d3", "yellow": "#e6daaa"
}
# Set plot defaults
sns.set_style(rc={"figure.facecolor": palette["dark"],
                  "axes.facecolor": palette["dark"],
                  "axes.edgecolor": palette["light"],
                  "grid.color": palette["grey"]})
```

```python
# Generate random sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
# Create and fit KMeans model
kmeans = KMeans(n_clusters=4, n_init="auto")
y_kmeans = kmeans.fit_predict(X)
```

```python
colours = [palette["blue"], palette["red"], palette["yellow"], palette["green"]]

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=[colours[label] for label in y_kmeans], s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c=palette["stone"], marker="x")

# sns.despine(left=True, bottom=True)

plt.title("K-Means Clustering Example", color=palette["light"])
plt.xlabel("Feature 1", color=palette["light"])
plt.ylabel("Feature 2", color=palette["light"])

# Customize ticks and labels
plt.tick_params(colors=palette["stone"], which='both')  # Set color for major and minor ticks
plt.show()
```
